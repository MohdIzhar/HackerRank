```python
# https://www.hackerrank.com/challenges/new-year-chaos/problem

# -> Time Limit Exceeded

#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the minimumBribes function below.
def minimumBribes(q):
    bribe = 0
    exceed = False
    for i in range(len(q)):
        mx = max(q)
        ind = q.index(mx)
        diff = (mx - 1) - ind
        if diff >= 0 and diff <= 2:
            q.remove(mx)
            bribe += diff
        elif diff > 2:
            exceed = True
            break

    if exceed:
        print("Too chaotic")
    else:
        print(bribe)
    


if __name__ == '__main__':
    t = int(input())

    for t_itr in range(t):
        n = int(input())

        q = list(map(int, input().rstrip().split()))

        minimumBribes(q)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the minimumBribes function below.
def minimumBribes(q):
    bribe = 0
    exceed = False
    for i in range(len(q) - 1, -1,-1):
        if i >= 2:
            if (q[i] < q[i-2]) and (q[i-2] - 1) == i and q[i-2] > q[i-1]:
                q[i-1], q[i-2] = q[i-2], q[i-1]
                q[i], q[i-1] = q[i-1], q[i]
                bribe += 2
            elif q[i] - 1 == i:
                continue
            elif q[i] < q[i-1] and q[i-1] - 1 == i:
                q[i], q[i-1] =  q[i-1], q[i]
                bribe += 1
            else:
                exceed = True
                break
            
        elif i < 2:
            if q[i] - 1 == i:
                continue
            elif q[i] < q[i-1] and q[i-1] - 1 == i:
                q[i], q[i-1] =  q[i-1], q[i]
                bribe += 1
            else:
                exceed = True
                break  

    if exceed:
        print("Too chaotic")
    else:
        print(bribe)

if __name__ == '__main__':
    t = int(input())

    for t_itr in range(t):
        n = int(input())

        q = list(map(int, input().rstrip().split()))

        minimumBribes(q)
```
