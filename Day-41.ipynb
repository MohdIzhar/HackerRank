```python
# https://www.hackerrank.com/challenges/class-1-dealing-with-complex-numbers/problem

import math

class Complex(object):
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary
        
    def __add__(self, no):
        self.r = self.real + no.real
        self.i = self.imaginary + no.imaginary
        if self.i == 0:
            result = "%.2f+0.00i" % (self.r)
        elif self.r == 0:
            if self.i >= 0:
                result = "0.00+%.2fi" % (self.i)
            else:
                result = "0.00-%.2fi" % (abs(self.i))
        elif self.i > 0:
            result = "%.2f+%.2fi" % (self.r, self.i)
        else:
            result = "%.2f-%.2fi" % (self.r, abs(self.i))
        return result

    def __sub__(self, no):
        self.r = self.real - no.real
        self.i = self.imaginary - no.imaginary
        if self.i == 0:
            result = "%.2f+0.00i" % (self.r)
        elif self.r == 0:
            if self.i >= 0:
                result = "0.00+%.2fi" % (self.i)
            else:
                result = "0.00-%.2fi" % (abs(self.i))
        elif self.i > 0:
            result = "%.2f+%.2fi" % (self.r, self.i)
        else:
            result = "%.2f-%.2fi" % (self.r, abs(self.i))
        return result

    def __mul__(self, no):
        self.r = self.real * no.real - self.imaginary * no.imaginary
        self.i = self.real * no.imaginary + self.imaginary * no.real
        if self.i == 0:
            result = "%.2f+0.00i" % (self.r)
        elif self.r == 0:
            if self.i >= 0:
                result = "0.00+%.2fi" % (self.i)
            else:
                result = "0.00-%.2fi" % (abs(self.i))
        elif self.i > 0:
            result = "%.2f+%.2fi" % (self.r, self.i)
        else:
            result = "%.2f-%.2fi" % (self.r, abs(self.i))
        return result

    def __truediv__(self, no):
        try:
            self.d = no.real * no.real + no.imaginary * no.imaginary
            self.r = (self.real * no.real + self.imaginary * no.imaginary) / self.d
            self.i = (no.real * self.imaginary - no.imaginary * self.real) / self.d
            if self.i == 0:
                result = "%.2f+0.00i" % (self.r)
            elif self.r == 0:
                if self.i >= 0:
                    result = "0.00+%.2fi" % (self.i)
                else:
                    result = "0.00-%.2fi" % (abs(self.i))
            elif self.i > 0:
                result = "%.2f+%.2fi" % (self.r, self.i)
            else:
                result = "%.2f-%.2fi" % (self.r, abs(self.i))
            return result

        except ZeroDivisionError:
            return None

    def mod(self):
        self.modu = math.sqrt(self.real * self.real + self.imaginary * self.imaginary)
        return "{0:.2f}+0.00i".format(self.modu)

    def __str__(self):
        if self.imaginary == 0:
            result = "%.2f+0.00i" % (self.real)
        elif self.real == 0:
            if self.imaginary >= 0:
                result = "0.00+%.2fi" % (self.imaginary)
            else:
                result = "0.00-%.2fi" % (abs(self.imaginary))
        elif self.imaginary > 0:
            result = "%.2f+%.2fi" % (self.real, self.imaginary)
        else:
            result = "%.2f-%.2fi" % (self.real, abs(self.imaginary))
        return result

if __name__ == '__main__':
    c = map(float, input().split())
    d = map(float, input().split())
    x = Complex(*c)
    y = Complex(*d)
    print(*map(str, [x+y, x-y, x*y, x/y, x.mod(), y.mod()]), sep='\n')
```

```python
# https://www.hackerrank.com/challenges/class-2-find-the-torsional-angle/problem

import math

class Points(object):
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, no):
        return Points((self.x - no.x), (self.y - no.y), (self.z - no.z))

    def dot(self, no):
        return (self.x * no.x) + (self.y * no.y) + (self.z * no.z)

    def cross(self, no):
        self.abx = self.y * no.z - self.z * no.y
        self.aby = self.z * no.x - self.x * no.z
        self.abz = self.x * no.y - self.y * no.x
        return Points(self.abx, self.aby, self.abz)
        
    def absolute(self):
        return pow((self.x ** 2 + self.y ** 2 + self.z ** 2), 0.5)

if __name__ == '__main__':
    points = list()
    for i in range(4):
        a = list(map(float, input().split()))
        points.append(a)

    a, b, c, d = Points(*points[0]), Points(*points[1]), Points(*points[2]), Points(*points[3])
    x = (b - a).cross(c - b)
    y = (c - b).cross(d - c)
    angle = math.acos(x.dot(y) / (x.absolute() * y.absolute()))

    print("%.2f" % math.degrees(angle))
```

```python
# https://www.hackerrank.com/challenges/angry-professor/problem

# Complete the angryProfessor function below.
def angryProfessor(k, a):
    cnt = 0
    for i in a:
        if i <= 0:
            cnt += 1

    if k == cnt or k < cnt:
        return "NO"
    else:
        return "YES"
```

```python
# https://www.hackerrank.com/challenges/beautiful-days-at-the-movies/problem

# Complete the beautifulDays function below.
def beautifulDays(i, j, k):
    movieon = 0
    for day in range(i,j+1):
        if abs(day - int(str(day)[::-1])) % k == 0:
            movieon += 1

    return movieon
```

```python
# https://www.hackerrank.com/challenges/minimum-swaps-2/problem

# Complete the minimumSwaps function below.
def minimumSwaps(arr):
    result = sorted(arr)
    index_dic = {v: i for i,v in enumerate(arr)}
    swaps = 0
    for i,v in enumerate(arr):
        correct = result[i]
        if v != correct:
            swap_index = index_dic[correct]
            arr[swap_index], arr[i] = arr[i], arr[swap_index]
            index_dic[v] = swap_index
            index_dic[correct] = i
            swaps += 1
    return swaps
```
